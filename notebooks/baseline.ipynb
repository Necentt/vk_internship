```python
drop_col = ['feature_0', 'feature_73', 'feature_74', 'feature_75']
train_df = train_df.drop(columns=drop_col)
test_df = test_df.drop(columns=drop_col)
```

```python
from catboost import CatBoostRanker, Pool

X_train = train_df.drop(columns=['search_id', 'target']).values
y_train = train_df['target'].values
queries_train = train_df['search_id'].values

X_test = test_df.drop(columns=['search_id', 'target']).values
y_test = test_df['target'].values
queries_test = test_df['search_id'].values
```

```python
train = Pool(
    data=X_train,
    label=y_train,
    group_id=queries_train
)

test = Pool(
    data=X_test,
    label=y_test,
    group_id=queries_test
)
```

```python
default_parameters = {
    'iterations': 2000,
    'custom_metric': ['NDCG', 'PFound', 'AverageGain:top=10'],
    'verbose': False,
    'random_seed': 42,
}

parameters = {}
```

```python
from copy import deepcopy

def fit_model(loss_function, additional_params=None, train_pool=train, test_pool=test):
    parameters = deepcopy(default_parameters)
    parameters['loss_function'] = loss_function
    parameters['train_dir'] = loss_function

    if additional_params is not None:
        parameters.update(additional_params)

    model = CatBoostRanker(**parameters)
    model.fit(train_pool, eval_set=test_pool, plot=True)

    return model
```

```python
model = fit_model('RMSE')
```

```python
test_df['predictions'] = model.predict(test)
```

```python
test_df.iloc[:, 76:].sort_values(by=['target', 'predictions'], ascending=False)
```

```python
from sklearn.metrics import ndcg_score

ndcg_score([np.asarray(y_test)], [np.asarray(test_df['predictions'])])
```
